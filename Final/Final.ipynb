```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("Airline.csv")
df
```

```python
#CLEANING DATASET
```

```python
df = df.dropna(how='any',axis=0) 
df
```

```python
#Encoding satisfaction
```

```python
df['satisfaction'] = df['satisfaction'].map({'satisfied': 1, 'dissatisfied': 0})
df
```

```python
#Encoding gender, Customer Type and Type of travel
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

df['Customer Type'] = df['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer': 0})

df['Type of Travel'] = df['Type of Travel'].map({'Personal Travel': 1, 'Business travel': 0})

df
```

```python
#Encoding Class
```

```python
df['Class'] = df['Class'].map({'Business': 2, 'Eco Plus': 1, 'Eco':0})

df
```

```python
df.drop(['Gate location'], axis=1, inplace=True)
df
```

# FIRST CLASSIFICATION: LOGISTIC

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PowerTransformer
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
#X/Y splitting
```

```python
X = df[['Gender', 'Customer Type', 'Age', 'Type of Travel',
       'Class', 'Flight Distance', 'Seat comfort',
       'Departure/Arrival time convenient', 'Food and drink',
       'Inflight wifi service', 'Inflight entertainment', 'Online support',
       'Ease of Online booking', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Cleanliness', 'Online boarding',
       'Departure Delay in Minutes', 'Arrival Delay in Minutes']]
y = df[['satisfaction']]
```

```python
# Train& test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=11)
```

```python
#SMOTE
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

```python
y_train_res.value_counts()
```

```python
count_data= sns.countplot(x='satisfaction', data=y_train_res)
count_data.set_xlabel("SATISFACTION", fontsize=12)
count_data.set_ylabel("COUNT", fontsize=12)
plt.show()
```

```python
lr = LogisticRegression() 

#YOU JUST FIT THE TRAIN. THE TEST IS FOR THE FINAL PART (WE DONT TOUCH IT UNTIL THE END).
lr.fit(X_train_res, y_train_res)
y_train_2 = lr.predict(X_train_res)
y_test_2 = lr.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_2, digits=2))
```

# SECOND CLASSIFICATION: DECISION TREES

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
```

```python

#from sklearn.metrics import plot_confusion_matrix



dt = DecisionTreeClassifier(max_depth=3)

dt.fit(X_train_res, y_train_res)

y_pred_train_dt = dt.predict(X_train_res)
y_pred_test_dt = dt.predict(X_test)


from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_test_dt , digits=2))
```

```python
from sklearn.tree import plot_tree
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (34,20))

plot_tree(dt,filled = True, rounded=True,feature_names=list(X.columns))
plt.show() 
```

```python
plt.figure(figsize=(20,14))
plt.barh(X.columns,dt.feature_importances_)
```

# THIRD CLASSIFICATION: RANDOM FORESTS

```python
from sklearn.ensemble import RandomForestClassifier

RAND_STATE = 42 # for reproducible shuffling
TT_RATIO = 0.2 # test/train

# min_samples_leaf: number of data per leaf.
# n_estimators: number of trees in the forest.
# bootstrap: True -> Create new dataset with samples with replacement. 
# Vamos que cojo muestras del dataset  para crear minidatasets y aplicar los trees.
# oob: out of bag

rfc_ops = {"max_depth":5,
           "min_samples_leaf":20,
           "n_estimators":100,
           "bootstrap":True,
           "oob_score":True,
           "random_state":RAND_STATE}

clf = RandomForestClassifier(**rfc_ops)

clf.fit(X_train_res, y_train_res)
print("train prediction accuracy score: %.2f" %(clf.score(X_train_res, y_train_res)))
print("test prediction accuracy score: %.2f"  %(clf.score(X_test, y_test)))
```

```python
clf.oob_score_
```

```python
plt.figure(figsize=(20,14))
plt.barh(X.columns,clf.feature_importances_)
```

# HYPER PARAMETERS TUNING

```python
#In this case we just need it for Random Forest
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [50, 100],
    'min_samples_split': [80],
    'min_samples_leaf' : [50],
    'max_features': ['sqrt'],
    ##'max_samples' : ['None', 0.5],
   'max_depth':[3,5]
    ## 'bootstrap':[True,False]
    }
clf = RandomForestClassifier(random_state=RAND_STATE)
```

```python
grid_search = GridSearchCV(clf, param_grid, cv=5,return_train_score=True,n_jobs=-1,)
grid_search.fit(X_train_res,y_train_res.values.ravel())
best_params = grid_search.best_params_ #To check the best set of parameters returned
best_params
```
